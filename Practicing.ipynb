```python
def sudoku_solver(board):

    def is_valid(row, col, num):
        # row
        for r in range(9):
            if board[row][r] == num:
                return False
        for c in range(9):
            if board[c][col] == num:
                return False

        box_row = (row//3)*3
        box_col = (col//3)*3

        for i in range(box_row, box_row+3):
            for j in range(box_col, box_col+3):
                if board[i][j]==num:
                    return False
        return True

    def backtrack():
        for row in range(9):

            for col in range(9):

                if board[row][col] == '.':

                    for num in '123456789':
                        if is_valid(row, col, num):

                            board[row][col] = num

                            if backtrack():
                                return True

                            # Backtrack
                            board[row][col] ='.'
                    return False
        return True

    backtrack()
    return board
```

```python
## power of 2

def power(x):
    if x<1:
        return False
    if x == 1:
        return True
    return power(x//2) and x%2 == 0

```

```python
def powoftwo(x):
    while x > 1:
        if x%2!=0:
            return False
        else:
            return True
        x = x//2

```

```python
def rev(string):
    index = len(string)-1

    if len(string) <=1:
        return string
    return string[index]+rev(string[:index])
```

```python
def is_palin(string):
    if len(string) <=1:
        return True
    if string[0] != string[-1]:
        return False
    return is_palin(string[1:-1])
```

```python
## with cache

fibs = {}

def fib_cache(n):
    if n<=1:
        return n
    if n in fibs:
        return fibs[n]
    fibs[n] = fib_cache(n-1) + fib_cache(n-2)
    return fibs[n]
```

```python

def flatten_list(val):
    l3 = []
    for i in val:
        if not isinstance(i, list):
            l3.append(i)
        else:
            l3.extend(flatten_list(i))
    return l3
```

```python
def bs_rec(nums, target, left = 0, right = None):
    if right is None:
        right = len(nums)-1

    if left > right:
        return -1
    mid = left + (right-left)//2
    if nums[mid] == target:
        return mid
    elif nums[mid]> target:
        return bs_rec(nums, target, left, mid-1)
    else:
        return bs_rec(nums, target, mid+1, right)
```

```python
def subsets(nums):
    res = []

    def helper(index, path):
        if index == len(nums):
            res.append(path[:])
            return

        #take
        path.append(nums[index])
        helper(index+1, path)

        # backtrack
        path.pop()
        #recurse
        helper(index+1, path)

    helper(0, [])
    return res
```

```python
subsets([1,2,3])
```

```python
def permutation(nums):
    res = []
    hashi = [False]* len(nums)

    def helper(path):
        if len(path) == len(nums):
            res.append(path[:])
            return

        for i in range(len(nums)):
            if hashi[i]:
                continue
            hashi[i] = True

            path.append(nums[i])
            helper(path)

            path.pop()
            hashi[i] = False
    helper([])
    return res
```

```python
permutation([1,2,3])
```
