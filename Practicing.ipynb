```python
def sudoku_solver(board):

    def is_valid(row, col, num):
        # row
        for r in range(9):
            if board[row][r] == num:
                return False
        for c in range(9):
            if board[c][col] == num:
                return False

        box_row = (row//3)*3
        box_col = (col//3)*3

        for i in range(box_row, box_row+3):
            for j in range(box_col, box_col+3):
                if board[i][j]==num:
                    return False
        return True

    def backtrack():
        for row in range(9):

            for col in range(9):

                if board[row][col] == '.':

                    for num in '123456789':
                        if is_valid(row, col, num):

                            board[row][col] = num

                            if backtrack():
                                return True

                            # Backtrack
                            board[row][col] ='.'
                    return False
        return True

    backtrack()
    return board
```

```python
## with cache

fibs = {}

def fib_cache(n):
    if n<=1:
        return n
    if n in fibs:
        return fibs[n]
    fibs[n] = fib_cache(n-1) + fib_cache(n-2)
    return fibs[n]
```

```python
def bs_rec(nums, target, left = 0, right = None):
    if right is None:
        right = len(nums)-1

    if left > right:
        return -1
    mid = left + (right-left)//2
    if nums[mid] == target:
        return mid
    elif nums[mid]> target:
        return bs_rec(nums, target, left, mid-1)
    else:
        return bs_rec(nums, target, mid+1, right)
```

```python
def subsets(nums):
    res = []

    def helper(index, path):
        if index == len(nums):
            res.append(path[:])
            return

        #take
        path.append(nums[index])
        helper(index+1, path)

        # backtrack
        path.pop()
        #recurse
        helper(index+1, path)

    helper(0, [])
    return res
```

```python
subsets([1,2,3])
```

```python
def permutation(nums):
    res = []
    hashi = [False]* len(nums)

    def helper(path):
        if len(path) == len(nums):
            res.append(path[:])
            return

        for i in range(len(nums)):
            if hashi[i]:
                continue
            hashi[i] = True

            path.append(nums[i])
            helper(path)

            path.pop()
            hashi[i] = False
    helper([])
    return res
```

```python
permutation([1,2,3])
```

```python
## N Queen

def n_queen(n):
    res = []
    board= [["."] * n for _ in range(n)]

    rows = set()
    pos_diag = set()
    neg_diag = set()

    def backtrack(col):
        if col == n:
            ans = ["".join(r) for r in board]
            res.append(ans)
            return

        for row in range(n):
            if ((row in rows) or
                    (row + col) in pos_diag or (
                            row - col) in neg_diag):
                continue

            rows.add(row)
            pos_diag.add(row+col)
            neg_diag.add(row-col)

            board[row][col] = 'Q'
            backtrack(col+1)

            rows.remove(row)
            pos_diag.remove(row+col)
            neg_diag.remove(row-col)
            board[row][col] = '.'

    backtrack(0)
    return res
```

```python
n_queen(7)
```

```python
def agg_cows(stalls, cows):
    stalls.sort()

    def is_possible(distance):
        last = stalls[0]
        count = 1
        for i in range(1, len(stalls)):
            if stalls[i] - last >= distance:
                count+=1
                last = stalls[i]
        return count>= cows

    left = 1; right = max(stalls)-min(stalls); ans = -1

    while left <= right:
        mid = left + (right-left)//2

        if is_possible(mid):
            ans = mid
            left = mid+1
        else:
            right = mid-1
    return ans
```

```python
print("agg cows: ", agg_cows([1,2,3,4,7], 1))
```
