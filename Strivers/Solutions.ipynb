```python
def print_n_to_1(n):
    if n <0:
        return
    print(n)
    print_n_to_1(n-1)
```

```python
print_n_to_1(10)
```

```python
def print_1_to_n(n):
    if n < 0:
        return
    print_1_to_n(n-1)
    print(n)
```

```python
print_1_to_n(10)
```

```python
def summation(n):
    if n<=0:
        return 0
    return n+summation(n-1)
```

```python
summation(3)
```
