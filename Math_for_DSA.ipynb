```python
def reverse_num(num):
    rev_num = 0
    while num:
        rev_num = rev_num*10 + num%10
        num //= 10
    return rev_num
```

```python
reverse_num(10400)
```

```python
def palindrome_check(num):
    rev_num = 0
    dup = num
    while num:
        rev_num = rev_num*10 + num%10
        num //= 10
    if rev_num == dup:
        return True
    return False

```

```python
palindrome_check(1231)
```

## Armstrong Number
An Armstrong number (or narcissistic number) is an \(n\)-digit integer that equals the sum of its own digits each raised to the power of \(n\). For example, \(153\) is a 3-digit Armstrong number because \(1^{3}+5^{3}+3^{3}=1+125+27=153\). Other examples include 0, 1, 370, 371, 407, and 1634.

```python
def is_armstrong_number(num):
    dup = num
    sum = 0
    n = len(str(num))
    while num>0:
        sum = sum + (num%10)**n
        num = num//10
    print(sum)
    if sum == dup:
        return True
    return False
```

```python
is_armstrong_number(1634)
```

```python
def find_divisors(num):
    l1 = []
    for i in range(1, num):
        if num%i==0:
            que = num//i
            l1.append(i)
            if i not in l1 and que not in l1:
                l1.append(num//i)
    return l1
```

```python
find_divisors(36)
```
