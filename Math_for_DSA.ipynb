```python
def reverse_num(num):
    rev_num = 0
    while num:
        rev_num = rev_num*10 + num%10
        num //= 10
    return rev_num
```

```python
reverse_num(10400)
```

```python
def palindrome_check(num):
    rev_num = 0
    dup = num
    while num:
        rev_num = rev_num*10 + num%10
        num //= 10
    if rev_num == dup:
        return True
    return False

```

```python
palindrome_check(1231)
```

## Armstrong Number
An Armstrong number (or narcissistic number) is an \(n\)-digit integer that equals the sum of its own digits each raised to the power of \(n\). For example, \(153\) is a 3-digit Armstrong number because \(1^{3}+5^{3}+3^{3}=1+125+27=153\). Other examples include 0, 1, 370, 371, 407, and 1634.

```python
def is_armstrong_number(num):
    dup = num
    sum = 0
    n = len(str(num))
    while num>0:
        sum = sum + (num%10)**n
        num = num//10
    print(sum)
    if sum == dup:
        return True
    return False
```

```python
is_armstrong_number(1634)
```

```python
def find_divisors(num):
    l1 = []
    count = 0
    for i in range(1, num):
        if num%i==0:
            que = num//i
            l1.append(i)
            if i ==que:
                break
            l1.append(que)
            count+=1
    return l1, count
```

```python
def find_divisors(num):
    from math import sqrt
    count = 0
    l1 = []
    for i in range(1, int(sqrt(num))):
        if num%i==0:
            l1.append(i)
            if num//i != i:
                l1.append(num//i)
                count+=1
    return l1, count
```

```python
find_divisors(36)
```

```python
def find_gcd(num1, num2):
    min_num = min(num1, num2)
    i = min(num1, num2)

    while i:
        if num1%i ==0 and num2%i==0:
            return i
        i-=1
```

## Euclidean GCD

for a and b; such that a>b

Subtraction version:
gcd(a,b) = gcd(a-b, b)

Division version:
gcd(a,b) = gcd(b, a mod b)
